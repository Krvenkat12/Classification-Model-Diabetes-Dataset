```python
proj_path = '/content/drive/MyDrive/CIS-545-Final-Project/Deliverable_1'
input_path = proj_path + '/input.ipynb'
ouput_path = proj_path + '/Outputs'

target_brfss_years = [2023, 2022, 2021, 2020, 2019, 2018, 2017, 2016, 2015] # Add targets Brfss years for modeling
```

```python
start_time = time.perf_counter()
for year in target_brfss_years:
  pm.execute_notebook(
    input_path,
    output_path = f'{ouput_path}/diabetes_df_{year}.ipynb',
    parameters={
        'year' : year
    }
  )
  end_time = time.perf_counter()
  elapsed_time = end_time - start_time

print(f"All notebooks executed sucessfuly in {elapsed_time//60}m {elapsed_time%60}s seconds")
```
